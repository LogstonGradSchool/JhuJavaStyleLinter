```python
df = pd.read_csv('results.edit.csv', header=None)
df.columns = ['repo', 'forks', 'stars', 'file', 'unorderedClasses', 'totalClasses']
df.head()
```

```python
df_agg = df.groupby(['repo', 'forks', 'stars']).sum()
df_agg['percentUnordered'] = (df_agg['unorderedClasses'] / df_agg['totalClasses']) * 100
# Filter any repos that did not have any classes.
df_agg = df_agg[df_agg['totalClasses'] > 0]
df_agg
```
